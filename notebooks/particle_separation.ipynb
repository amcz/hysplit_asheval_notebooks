Looks at particle separation between 0.6, 6 and 20 um particles.

Runs are in directory RunB/control/
- b0 diameter 6um
- b1 diameter 20 um
- b2 diamter 0.6 um

```python
%load_ext autoreload
```

```python
%autoreload
import numpy as np
import datetime
import matplotlib.pyplot as plt

import part_sep as ke

from monetio.models import hysplit
from monetio.models import pardump
from utilhysplit import par2conc
import feature_tracking
```

```python
kas1 = ke.BezyExample()
```

```python
%autoreload
# Create fits to the smallest particle size
# Adjust number of components if number of particles is too small.
pdumpb2 = kas1.get_pdump("b2",verbose=False)
lev=[1000,20000]
poll=1
nnn=10
method='gmm'
ft2 = feature_tracking.FeatureTracker(method,nnn,poll,lev)
ft2.create(pdumpb2)
print('done')
```

```python
%autoreload
# Create fits to the smallest particle size
# Adjust number of components if number of particles is too small.
pdumpb1 = kas1.get_pdump("b1",verbose=False)
lev=[1000,20000]
poll=1
nnn=10
method='gmm'
ft1 = feature_tracking.FeatureTracker(method,nnn,poll,lev)
ft1.create(pdumpb1)
print('done')
```

```python
%autoreload
pdumpb0 = kas1.get_pdump("b0",verbose=False)
lev=[1000,20000]
poll=1
nnn=10
method='gmm'
ft0 = feature_tracking.FeatureTracker(method,nnn,poll,lev)
ft0.create(pdumpb0)
print('done')
```

```python
ft1.plotscore()
```

```python
ft0.plotscore()
```

```python
# Compare scores for 0.6um and 6um particle sizes.
# These remain close together for the entire time period.
%autoreload
score = feature_tracking.compare(ft0,ft2,name1='p6', name2='p0.6')
ax = plt.gca()
d1 = datetime.datetime(2020,10,21,20)
d2 = datetime.datetime(2020,10,22,15)
ax.set_xlim(d1,d2)
ax.set_ylim(-5,1)
plt.tight_layout()
plt.savefig('particles_size_comparison_02.png')
plt.show()
```

```python
# Compare scores for 6um and 20um particle sizes.
# These separate.
%autoreload
score = feature_tracking.compare(ft0,ft1,name1='p6', name2='p20')
d1 = datetime.datetime(2020,10,21,20)
d2 = datetime.datetime(2020,10,22,15)
ax = plt.gca()
ax.set_xlim(d1,d2)
ax.set_ylim(-10,1)
plt.tight_layout()
plt.savefig('particles_size_comparison_10.png')
plt.show()
```
